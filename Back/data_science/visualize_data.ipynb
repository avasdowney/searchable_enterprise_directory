```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read the data

df = pd.read_csv('data.csv')
df.head()
```

# Distribution of Data

```python
# graph distribution of job roles

graph = sns.countplot(y='job_role', data=df)
graph.set(xlabel ="Count", ylabel = "Job Role", title ='Employee Count by Job Role')
```

```python
# graph distribution of job locations

graph = sns.countplot(y='work_location', data=df)
graph.set(xlabel ="Count", ylabel = "Job Location", title ='Employee Count by Job Location')
```

```python
# remove CEO from graph because hes an outlier
df2 = df.iloc[1: , :]
```

```python
# Scatterplot showing salary versus predictors

plt.figure(figsize=[6,9])
graph = sns.scatterplot(x="work_location", y="salary",
                        hue="job_role", data=df2)
graph.set(xlabel ="Work Location", ylabel = "Salary ($USD)", title ='Employee Salary by Job Location and Role')
```
